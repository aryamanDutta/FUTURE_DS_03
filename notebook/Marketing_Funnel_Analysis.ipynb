```
# ==========================
# Marketing Funnel Stage Counts
# ==========================

total_visitors = len(df)

viewed = df["viewed_product"].sum()

cart = df["added_to_cart"].sum()

checkout = df["checkout_started"].sum()

purchased = df["purchase_completed"].sum()

funnel_df = pd.DataFrame({
    "Stage": [
        "Visitors",
        "Viewed Product",
        "Added to Cart",
        "Checkout Started",
        "Purchase Completed"
    ],
    "Users": [
        total_visitors,
        viewed,
        cart,
        checkout,
        purchased
    ]
})

funnel_df
```

```
plt.figure(figsize=(8,5))

ax = sns.barplot(
    data=funnel_df,
    x="Stage",
    y="Users",
    palette="viridis"
)

for container in ax.containers:
    ax.bar_label(container, fmt='%d')

plt.title("Marketing Funnel")
plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

```
# ==========================
# Conversion Rates
# ==========================

visitor_to_view = viewed / total_visitors * 100

view_to_cart = cart / viewed * 100

cart_to_checkout = checkout / cart * 100

checkout_to_purchase = purchased / checkout * 100

overall_conversion = purchased / total_visitors * 100

conversion_df = pd.DataFrame({
    "Stage":[
        "Visitor → Product",
        "Product → Cart",
        "Cart → Checkout",
        "Checkout → Purchase",
        "Overall Conversion"
    ],
    "Conversion Rate (%)":[
        visitor_to_view,
        view_to_cart,
        cart_to_checkout,
        checkout_to_purchase,
        overall_conversion
    ]
})

conversion_df.round(2)
```

```
plt.figure(figsize=(9,5))

ax = sns.barplot(
    data=conversion_df,
    x="Stage",
    y="Conversion Rate (%)",
    palette="Set2"
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%")

plt.title("Conversion Rates Across Funnel Stages")

plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

```
dropoffs = pd.DataFrame({

    "Stage":[
        "Visitor",
        "Product View",
        "Cart",
        "Checkout"
    ],

    "Users Lost":[
        total_visitors-viewed,
        viewed-cart,
        cart-checkout,
        checkout-purchased
    ]
})

dropoffs
```

```
plt.figure(figsize=(8,5))

ax = sns.barplot(
    data=dropoffs,
    x="Stage",
    y="Users Lost",
    palette="Reds"
)

for container in ax.containers:
    ax.bar_label(container, fmt="%d")

plt.title("User Drop-Off at Each Funnel Stage")

plt.tight_layout()

plt.show()
```

# Business Insights & Recommendations

## Key Findings

- The dataset contains 120,000 visitors with an overall conversion rate of 6.82%.

- The largest customer drop-off occurs between the Product View and Add to Cart stages, where over 50,000 users leave the funnel.

- Paid Ads generate the highest overall revenue despite having a slightly lower conversion rate than Email.

- Email marketing achieves the highest purchase conversion rate (7.31%), indicating high-quality traffic.

- Mobile users generate significantly more revenue due to higher traffic volume, while Desktop users show a slightly higher conversion rate.

- Metro regions contribute a larger share of total revenue compared to Non-Metro regions.

- Discount campaigns generate the highest total revenue.

- Products with discounts show a much higher purchase completion rate, indicating discounts positively influence conversions.

## Recommendations

- Improve the product page and cart experience to reduce the largest drop-off between Product View and Add to Cart.

- Increase investment in Email campaigns because they deliver the highest conversion efficiency.

- Continue using Paid Ads for scale but optimize targeting to improve conversion.

- Prioritize mobile optimization since most users access the platform using mobile devices.

- Focus future marketing campaigns on Metro regions where revenue contribution is highest.

- Use discounts strategically during high-value campaigns to maximize purchases while protecting profit margins.
